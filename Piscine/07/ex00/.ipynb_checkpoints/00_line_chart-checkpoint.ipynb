## create a connection to the database using the library sqlite3

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('data/checking-logs.sqlite.sqlite copy')
pd.io.sql.read_sql('select * from pageviews', conn)
```

## run a query that gets the datetime from the pageviews table, selecting only the users and not the admins using Pandas &&& create a new dataframe where the visits are counted and grouped by date

```python
visits = pd.io.sql.read_sql("""
SELECT strftime("%Y-%m-%d",datetime) AS datetime, COUNT(strftime("%Y-%m-%d",datetime)) AS views
    FROM pageviews
    WHERE uid LIKE "user_%"
    GROUP BY strftime("%Y-%m-%d",datetime)
""", conn)
visits
```

```python
visits.plot(figsize=(15,8),  fontsize=8, x='datetime', rot=90)
```
