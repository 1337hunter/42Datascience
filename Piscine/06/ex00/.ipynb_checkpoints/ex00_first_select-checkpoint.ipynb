## create a connection to the database using the library sqlite3

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('data/checking-logs.sqlite.sqlite copy')
```

## get the schema of the table pageviews

```python
pd.io.sql.read_sql("PRAGMA table_info(pageviews);", conn)
```

## get only the first 10 rows of the table pageviews to check what the table looks like

```python
pd.io.sql.read_sql("select * from pageviews limit 10", conn)
```

## get the subtable using only one query where

```python
pd.io.sql.read_sql("select uid, datetime from pageviews limit 10", conn)
```

```python
pd.io.sql.read_sql("select * from pageviews where uid  like 'user_%'", conn)
```

```python
pd.io.sql.read_sql("select * from pageviews order by uid", conn)
```

```python
pd.io.sql.read_sql("select * from pageviews", conn, index_col='datetime')
```

```python
pd.io.sql.read_sql("select * from pageviews", conn, index_col='datetime', parse_dates={'datetime':'%Y%m%d %H:%M:%S'})
```

```python
pageviews = pd.io.sql.read_sql("select * from pageviews", conn, index_col='datetime', params={"name" : 'pageviews'})
```
