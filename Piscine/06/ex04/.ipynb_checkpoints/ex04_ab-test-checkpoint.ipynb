## create a connection to the database using the library sqlite3

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('../ex00/data/checking-logs.sqlite.sqlite copy')
```

## using only one query for each of the groups, create two dataframes: test_results and control_results with the columns time and avg_diff and only two rows

```python
test_results = pd.io.sql.read_sql("""
SELECT 'after' AS time, avg(tab2.after) AS avg_diff
FROM
    (SELECT uid,
            avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS before
    FROM (
            SELECT  test.*,
                    (
                        SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE test.labname = deadlines.labs
                    ) AS deadlines_labs
            FROM    test
            WHERE   labname NOT IN ('project1') AND
                    julianday(first_commit_ts) < julianday(first_view_ts))
    GROUP BY uid)
    AS      tab1
    INNER JOIN
    (
        SELECT  uid, after
        FROM
            (SELECT uid,
                    avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS after
            FROM (
                    SELECT  test.*,
                        (SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE test.labname = deadlines.labs) AS deadlines_labs
                    FROM    test
            WHERE
                labname NOT LIKE 'project1' AND
                julianday(first_commit_ts) >= julianday(first_view_ts))
            GROUP BY uid)
            ) tab2
ON  tab1.uid = tab2.uid
UNION
SELECT  'before' AS time, avg(tab1.before) AS avg_diff
FROM
    (SELECT uid,
            avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS before
    FROM (
            SELECT  test.*,
                    (
                        SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE test.labname = deadlines.labs
                    ) AS deadlines_labs
            FROM    test
            WHERE   labname NOT IN ('project1') AND
                    julianday(first_commit_ts) < julianday(first_view_ts))
    GROUP BY uid)
    AS      tab1
    INNER JOIN
    (
        SELECT  uid, after
        FROM
            (SELECT uid,
                    avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS after
            FROM (
                    SELECT  test.*,
                        (SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE test.labname = deadlines.labs) AS deadlines_labs
                    FROM    test
            WHERE
                labname NOT LIKE 'project1' AND
                julianday(first_commit_ts) >= julianday(first_view_ts))
            GROUP BY uid)
            ) tab2
ON  tab1.uid = tab2.uid
""", conn)
test_results
```

```python
control_results = pd.io.sql.read_sql("""
SELECT 'after' AS time, avg(tab2.after) AS avg_diff
FROM
    (SELECT uid,
            avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS before
    FROM (
            SELECT  control.*,
                    (
                        SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE control.labname = deadlines.labs
                    ) AS deadlines_labs
            FROM    control
            WHERE   labname NOT IN ('project1') AND
                    julianday(first_commit_ts) < julianday(first_view_ts))
    GROUP BY uid)
    AS      tab1
    INNER JOIN
    (
        SELECT  uid, after
        FROM
            (SELECT uid,
                    avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS after
            FROM (
                    SELECT  control.*,
                        (SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE control.labname = deadlines.labs) AS deadlines_labs
                    FROM    control
            WHERE
                labname NOT LIKE 'project1' AND
                julianday(first_commit_ts) >= julianday(first_view_ts))
            GROUP BY uid)
            ) tab2
ON  tab1.uid = tab2.uid
UNION
SELECT  'before' AS time, avg(tab1.before) AS avg_diff
FROM
    (SELECT uid,
            avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS before
    FROM (
            SELECT  control.*,
                    (
                        SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE control.labname = deadlines.labs
                    ) AS deadlines_labs
            FROM    control
            WHERE   labname NOT IN ('project1') AND
                    julianday(first_commit_ts) < julianday(first_view_ts))
    GROUP BY uid)
    AS      tab1
    INNER JOIN
    (
        SELECT  uid, after
        FROM
            (SELECT uid,
                    avg((julianday(deadlines_labs) - julianday(first_commit_ts)) * 24) AS after
            FROM (
                    SELECT  control.*,
                        (SELECT datetime(deadlines, 'unixepoch')
                        FROM deadlines
                        WHERE control.labname = deadlines.labs) AS deadlines_labs
                    FROM    control
            WHERE
                labname NOT LIKE 'project1' AND
                julianday(first_commit_ts) >= julianday(first_view_ts))
            GROUP BY uid)
            ) tab2
ON  tab1.uid = tab2.uid
""", conn)
control_results
```
